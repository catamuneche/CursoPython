```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
casas.head()
```

```python
casas.columns
```

```python
X = casas[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
```

```python
X
```

```python
Y = casas['Price']
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
lrm = LinearRegression() #MODELO DE ENTRENAMIENTO
lrm.fit(X_train, Y_train)
```

```python
##Evaluar el modelo (modelo de prueba)
```

```python
from sklearn import metrics
```

```python
predicciones = lrm.predict(X_test)
predicciones
```

```python
Y_test
```

```python
plt.scatter(Y_test, predicciones)
```

```python
sns.distplot(Y_test - predicciones)
```

MAE (Mean Absolute error) - Media del valor absoluto de errores

```python
metrics.mean_absolute_error(Y_test,predicciones)
```

MSE (medio de los errores al cuadrado)

```python
metrics.mean_squared_error(Y_test,predicciones)
```

RMSE (raiz cuadrada de la media de los errores al cuadrado)

```python
np.sqrt(metrics.mean_squared_error(Y_test,predicciones))
```
