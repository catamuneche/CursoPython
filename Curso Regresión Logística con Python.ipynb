```python
def edad_media(columnas):
    edad=columnas[0]
    clase=columnas[1]
    if pd.isnull(edad):
        if clase == 1:
            return 38
        elif clase == 2:
            return 30
        else:
            return 25
    else:
        return edad
```

```python
entrenamiento['Age']= entrenamiento[['Age','Pclass']].apply(edad_media,axis=1)
```

```python
entrenamiento.isnull()
sns.heatmap(entrenamiento.isnull())
```

```python
entrenamiento.drop('Cabin',axis=1,inplace=True)
```
