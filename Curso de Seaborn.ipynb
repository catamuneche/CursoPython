# Mapas de calor (Heatmaps)

```python
import seaborn as sns
```

```python
vuelos = sns.load_dataset('flights')
vuelos.head(10)
```

```python
vuelos_matrix=vuelos.pivot_table(index='month',columns='year',values='passengers')
vuelos_matrix
```

```python
sns.heatmap(vuelos_matrix)
```

```python
sns.heatmap(vuelos_matrix,cmap='plasma')
#'plasma' 'viridis' 'inferno' 'magma' 'coolwarm'
```

```python
sns.heatmap(vuelos_matrix,cmap='coolwarm')
```

```python
sns.heatmap(vuelos_matrix,cmap='plasma',linecolor='w',linewidths=2)
```

# Grficos de cuadrcula (grids): Pairgrid y FacetGrid

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flores = sns.load_dataset('iris')
flores.head()
```

```python
sns.pairplot(flores)
```

```python
graficos = sns.PairGrid(flores)
graficos.map(plt.scatter) 
```

```python
graficos = sns.PairGrid(flores)
graficos.map_diag(sns.distplot) 
graficos.map_upper(plt.scatter)
```

```python
graficos = sns.PairGrid(flores)
graficos.map_diag(sns.distplot) 
graficos.map_upper(plt.scatter)
graficos.map_lower(sns.kdeplot)
```

# Estilos y colores

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
propinas = sns.load_dataset('tips')
propinas.head()
```

```python
sns.countplot(x='sex', data=propinas)
```

```python
sns.set_style('darkgrid')
sns.countplot(x='sex', data=propinas)
sns.despine()
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex', data=propinas)
sns.despine()
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex', data=propinas)
sns.despine(left=True,bottom=True)
```

```python
plt.figure(figsize=(12,4))
sns.set_style('whitegrid')
sns.countplot(x='sex', data=propinas)
sns.despine(left=True,bottom=True)
```

```python
plt.figure(figsize=(8,4))
sns.set_style('whitegrid')
sns.countplot(x='sex', data=propinas)
sns.despine(left=True,bottom=True)
```

```python
plt.figure(figsize=(1,4))
sns.set_style('whitegrid')
sns.countplot(x='sex', data=propinas)
sns.despine(left=True,bottom=True)
```

```python
sns.set_context(font_scale=1.5)
plt.figure(figsize=(8,4))
sns.set_style('whitegrid')
sns.countplot(x='sex', data=propinas)
sns.despine(left=True,bottom=True)
```

```python
sns.set_context('notebook',font_scale=1.5)
plt.figure(figsize=(8,4))
sns.set_style('whitegrid')
sns.countplot(x='sex', data=propinas)
sns.despine(left=True,bottom=True)
```
